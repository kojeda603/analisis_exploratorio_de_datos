```python
columnas_numericas = df.select_dtypes([float,int]).columns
print('columnas_numericas')
print(columnas_numericas)
```

```python
columnas_categoricas = df.select_dtypes([object]).columns
print('columnas_categoricas')
print(columnas_categoricas)
```

```python
df[columnas_categoricas] = df[columnas_categoricas].astype('category')
```

```python
df.info()
```

```python
for col in columnas_categoricas:
  print("")
  print(col)
  print(df[col].cat.categories)
```

```python
# estadistica descriptiva
print("\n"+"="*60)
print("estadistica descriptiva")
print("="*60)

columnas_numericas
df[columnas_numericas].describe().round(2)

print("\nDistrucion por sexo:")
print(df['sex'].value_counts())
print(f"\nPorcentaje:")
print((df["sex"].value_counts(normalize=True)*100).round(2))

print("\nDistribucion por año: ")
print(df["year"].value_counts().sort_index())

print("\nDistribucion por estado de fumador")
print(df['smoker'].value_counts())
```

```python
df.isnull().sum()
```

```python
print("\n"+"="*60)
print("analisis de valores faltantes")
print("="*60)

missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
missing_regicor = pd.DataFrame({'Faltantes': missing, 'Porcentaje': missing_pct})
missing_regicor = missing_regicor[missing_regicor['Faltantes'] > 0].sort_values('Porcentaje', ascending=False)
print("\nValores faltantes por columna")
missing_regicor
```

```python
# 4. analisis por gripos

print("\n" + "=" * 60)
print("analisis por grupos")
print("=" * 60)

print("\nEstadisticas por sexo")
print(df.groupby("sex")[columnas_numericas].mean().round(2))

print("\nEstadisticas por año")
print(df.groupby("year")[columnas_numericas[2:6]].mean().round(2))

print("\nEstadisticas por estado de fumador:")
print(df.groupby("smoker")[columnas_numericas[2:6]].mean().round(2))
```

```python
pd.crosstab(df['sex'], df['smoker'], margins=True)
```

```python
import seaborn as sns

sns.heatmap(pd.crosstab(df['smoker'],df['sex']), annot=True, cmap="YlGnBu", fmt='d')
plt.show()
```

```python
sns.catplot(data=df.query('year == 2005'), x='sex', kind='count', col='smoker', hue='sex', palette='husl')
plt.show()
```

```python
sns.catplot(data=df.query('year == 2005'), x='sex', y='chol', kind='box', hue='sex', palette='husl',col='histchol')
plt.show()
```
