```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.shape
```

```python
penguins.columns
```

```python
penguins["body_mass_g"].hist()
plt.show()
```

```python
sns.pairplot(data=penguins)
plt.show()
```

```python
penguins.columns
```

```python
sns.pairplot(data=penguins, hue="island")
plt.show()
```

```python
sns.pairplot(data=penguins, hue="species")
plt.show()
```

```python
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species",hue_order=["Gentoo","Adelie","Chinstrap"])
plt.show()
```

```python
# Seleccionar solo columnas numricas
numeric_cols = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']

# Crear heatmap de correlacin
plt.figure(figsize=(10, 8))
correlation = penguins[numeric_cols].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f',
            linewidths=1, square=True, cbar_kws={'label': 'Correlacin'})
plt.title('Correlacin entre Medidas Fsicas de Pinginos')
plt.tight_layout()
plt.show()
```

```python
sns.relplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species", kind="line", col="island", row="sex")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.countplot(data=penguins, x="species", hue="species", palette="colorblind")
plt.show()
```

```python
sns.scatterplot(data=penguins, x="flipper_length_mm", y="bill_depth_mm", hue="species", style="species", palette="colorblind", size="body_mass_g")

plt.show()
```

```python
penguins["species"].value_counts(normalize=True).plot(kind="bar")
```

```python
sns.barplot(data=penguins, x="species", y="bill_length_mm", hue="species", palette="colorblind")
plt.show()
```

```python
sns.boxplot(data=penguins, x="species", y="bill_length_mm", hue="sex", hue_order=["Female","Male"], palette="husl")
plt.show()
```
