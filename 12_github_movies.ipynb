```python
import pandas as pd
```

```python
movie = pd.read_csv('https://raw.githubusercontent.com/Gustavo-Bruges/movies/refs/heads/main/movies.csv')
rating = pd.read_csv('https://raw.githubusercontent.com/Gustavo-Bruges/movies/refs/heads/main/ratings.csv')
```

```python
print(movie.shape)
print(rating.shape)
```

```python
print("-"*100)
print("movie dataset")
print(movie.head())

print("-"*100)
print("rating dataset")
print(rating.head())
```

```python
movie_ratings = movie.merge(rating, on='id', suffixes=('_movie', '_rating'))
movie_ratings.head()
```

```python
movie_ratings.drop(["Unnamed: 0_movie",'Unnamed: 0_rating'], axis=1, inplace=True)
```

```python
movie_ratings.head()
```

```python
print(movie_ratings.shape)
```

```python
movie_genres = pd.read_csv('https://raw.githubusercontent.com/Gustavo-Bruges/movies/refs/heads/main/movie_to_genres.csv')
```

```python
movie_genres.head()
```

```python
movie_genres.drop(["Unnamed: 0"], axis=1, inplace=True)
print(movie_genres.head())
```

```python
movie_generos = movie.merge(movie_genres, left_on='id', right_on=('movie_id'))
movie_generos.head()
```

```python
sci_movie = movie_generos[movie_generos['genre'] == 'Science Fiction']
print(sci_movie.head())
print(sci_movie.shape)
```

```python
action_movie = movie_generos[movie_generos['genre'] == 'Action']
print(action_movie.head())
print(action_movie.shape)
```

```python
df_left = action_movie.merge(sci_movie, on='movie_id', how='left', suffixes=('_action', '_sci'))

print(df_left.shape)
```

```python
df_left.head()
```

```python
df_right = action_movie.merge(sci_movie, on='movie_id', how='right', suffixes=('_action', '_sci'))
```
