```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://github.com/Gustavo-Bruges/ciencia-de-datos-unab/raw/refs/heads/datos/olympics.zip")
```

#Bloque 1. Exploración estructural

## 1.	¿Cuántas filas y columnas tiene el dataset?

```python
filas,columnas = df.shape
print(f"La data frame netflix tiene {filas} filas y tiene {columnas} columnas")
```

## 2.	¿Cuáles son las variables categóricas?

```python
df.select_dtypes(include=['object']).columns
```

## 3.	¿Cuántos deportes distintos hay en la columna sport?

```python
print(f"Existen {df['sport'].nunique()} deportes distintos")
print(df['sport'].unique())
```

4.	¿Cuántos países distintos aparecen en la variable team?

```python
print(f"Existen {df['team'].nunique()} países distintos")
print(df['team'].unique())
```

# Bloque 2. Filtrado

## 5.	Filtra los registros donde el atleta sea mujer (sex == "F").

```python
df_mujeres = df[df['sex'] == 'F']
df_mujeres
```

## 6.	Filtra los registros donde se haya ganado una medalla.

```python
df_medalla = df[df['medal'].notna()]
df_medalla
```

## 7.	¿Cuántas medallas se registran en los Juegos Olímpicos de 2012?

```python
total = df[df['year'] == 2012]['medal'].notna().sum()
print(f"Medallas en los Juegos Olímpicos de 2012: {total}")
```

# Bloque 3. Agregación

## 8.	Calcula el número total de medallas por país (team).

```python
medallas_por_pais = df[df['medal'].notna()].groupby('team')['medal'].count().sort_values(ascending=False)
print(medallas_por_pais)
```

## 9.	¿Cuál es el país con mayor número de medallas?

```python
print(f"El país con mayor número de medallas es {medallas_por_pais.idxmax()} con {medallas_por_pais.max()}")
```

## 10.	Calcula el número de medallas por tipo (Gold, Silver, Bronze).

```python
medallas_por_tipo = df[df['medal'].notna()].groupby('medal')['medal'].count().sort_values(ascending=False)
print(medallas_por_tipo)
```

# Bloque 4. Visualización básica

## 11.	Construye un gráfico de barras de el número de medallas por tipo


```python
medallas_por_tipo = df[df['medal'].notna()]['medal'].value_counts().reset_index()

sns.barplot(data=medallas_por_tipo, x='medal', y='count')
plt.title('Número de medallas por tipo')
plt.xlabel('Tipo de medalla')
plt.ylabel('Total')
plt.show()
```

## 12.	Construye un gráfico de barras de los 10 países con más medallas.


```python
plt.figure(figsize=(12, 6))
sns.barplot(data=medallas_por_pais.head(10).reset_index(), x='team', y='medal')
plt.title('Top 10 países con más medallas')
plt.xlabel('País')
plt.ylabel('Total de medallas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Bloque 5. Análisis por sexo

## 13.	Calcula el número de medallas ganadas por hombres y por mujeres.

```python
medallas_por_sexo = df[df['medal'].notna()]['sex'].value_counts().reset_index()
print(medallas_por_sexo)
```

## 14.	Construye un gráfico de barras comparando medallas por sexo.

```python
sns.barplot(data=medallas_por_sexo, x='sex', y='count')
plt.title('Medallas por sexo')
plt.xlabel('Sexo')
plt.ylabel('Total de medallas')
plt.show()
```

# Pregunta integradora

## 15.	¿Se observa una diferencia importante en el número de medallas entre hombres y mujeres?, justifica con el gráfico.


```python
total_m = medallas_por_sexo[medallas_por_sexo['sex'] == 'M']['count'].values[0]
total_f = medallas_por_sexo[medallas_por_sexo['sex'] == 'F']['count'].values[0]
diferencia = ((total_m - total_f) / total_f) * 100

print(f"Los hombres ganaron {total_m} medallas y las mujeres {total_f}, lo que significa que los hombres ganaron un {diferencia:.1f}% más medallas que las mujeres.")
```

# Reto opcional

## 16.	Calcula el promedio de edad de los atletas que ganaron medalla de oro.

```python
promedio_edad = df[df['medal'] == 'Gold']['age'].mean()
print(f"El promedio de edad de los atletas que ganaron medalla de oro es: {promedio_edad:.1f} años")
```

# Analisis extras

```python
# Filtrar solo medallas de oro en Tenis de Mesa del 1900 al 2016
df_gold_tt = df[(df['medal'] == 'Gold') &
                 (df['sport'] == 'Table Tennis') &
                   (df['year'].between(1900, 2016))]

# Crear gráfica ampliada
plt.figure(figsize=(10, 6))
sns.countplot(data=df_gold_tt, y='team', order=df_gold_tt['team'].value_counts().index)
plt.xlabel('Cantidad de Medallas de Oro ')
plt.ylabel('País')
plt.title('Medallas de Oro en Tenis de Mesa por País (1900-2016)')
plt.tight_layout()
plt.show()
```
