```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/ciencia_de_datos/Salud_Mental_20260406.csv")
```

```python

grupos_edad = ['Menor a 1', 'De 1 a 4', 'De 5 a 9', 'De 10 a 14',
               'De 15 a 19', 'De 20 a 49', 'De 50 a 64', '65 Y MAS']

df_agrupado = df.groupby('diagnostico_ingreso')[grupos_edad].sum()

# Total de casos por grupo de edad (suma de todos los diagnósticos)
total_por_grupo = df_agrupado.sum()

prevalencia_por_edad = pd.DataFrame({
    'Grupo de edad': grupos_edad,
    'Diagnóstico más prevalente': [df_agrupado[col].idxmax() for col in grupos_edad],
    'Casos': [df_agrupado[col].max() for col in grupos_edad],
    'Total grupo': [total_por_grupo[col] for col in grupos_edad],
})

prevalencia_por_edad['% del grupo'] = (
    prevalencia_por_edad['Casos'] / prevalencia_por_edad['Total grupo'] * 100
).round(1).astype(str) + '%'

print(tabulate(prevalencia_por_edad,
               headers='keys',
               tablefmt='pretty',
               showindex=False))
```
