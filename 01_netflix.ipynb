max y min duracion para USA en el genero de horror

```python
df_usa_movies = df_movies[df_movies['country']=='United States']
```

```python
# Max y min directamente
max_duration = df_usa_movies['duration'].max()
min_duration = df_usa_movies['duration'].min()

print(f"Duracin mxima: {max_duration} minutos")
print(f"Duracin mnima: {min_duration} minutos")
```

```python
max_title = df_usa_movies[df_usa_movies['duration'] == max_duration]['title'].values[0]
min_title = df_usa_movies[df_usa_movies['duration'] == min_duration]['title'].values[0]

print(f"\nPelcula ms larga: {max_title} con {max_duration} minutos")
print(f"Pelcula ms corta: {min_title} con {min_duration} minutos")
```
